## Download NetCDF files

```python
# Import system and functions

import sys
sys.path.append(r'C:\Users\anda4748\OneDrive\04_Processing\00_PythonScripts\00_My_Functions')

import os
import glob
import xarray as xr
from natsort import natsorted

from download_functions2 import (
    download_netcdf_ereefs,
    download_netcdf_noaa,
    download_netcdf_csiro
)

# Ensure the download directory exists
download_dir = r'X:\08_Hydrodynamics\CSIRO_CAWCR_Waves\CSIRO-CAWCR_wave-partitions_7kmgrid\Australia/'


# Date Format String Examples:

    # %Y-%m-%d   ==> 2024-04-15	
    # %Y-%m	     ==> 2024-04	
    # %Y%m%d     ==> 20240415	
    # %b-%Y	     ==> Apr-2024
    # %Y%m	     ==> 202404	
    # %Y_%m_%d	 ==> 2024_04_15	
    # %d-%b-%Y	 ==> 15-Apr-2024	
    # %Y%m%d%H%M ==> 202404151330	
```

```python
# Download eReefs files

download_netcdf_ereefs(
    base_url='https://thredds.ereefs.aims.gov.au/thredds/fileServer/ereefs/gbr4_v2/daily-monthly/',
    download_dir=download_dir,
    start_year=2010, start_month=9,
    end_year=2024, end_month=1,
    file_prefix='EREEFS_AIMS-CSIRO_gbr4_v2_hydro_daily-monthly-',
    date_format='%Y-%m' # Adjust if needed
)
```

```python
# Download NOAA files

download_netcdf_noaa(
    base_url='https://www.star.nesdis.noaa.gov/pub/socd/mecb/crw/data/5km/v3.1_op/nc/v1.0/daily/dhw/',
    download_dir=download_dir,
    start_year=2023, start_month=1,
    end_year=2025, end_month=1,
    file_prefix='ct5km_dhw_v3.1_',
    date_format='%Y%m%d'  # Adjust if needed
)
```

```python
# Download CAWCR-CSIRO files

download_netcdf_csiro(
    base_url="https://data-cbr.csiro.au/thredds/fileServer/catch_all/CMAR_CAWCR-Wave_archive/CAWCR_Wave_Hindcast_aggregate/gridded/",
    download_dir=download_dir,
    start_year=2025, start_month=3,
    end_year=2025, end_month=9,
    file_prefix="ww3.aus_4m.",
    date_format="%Y%m"
)
```

```python
# List all files with .nc extension
data_files = natsorted(glob.glob(os.path.join(download_dir, '*.nc')))

# Check if files were found
if data_files:
    print(f"Found {len(data_files)} NetCDF files.")
    # Open the first file
    data = xr.open_dataset(data_files[0], engine='netcdf4', decode_cf=False)
    print(data)
else:
    print("No NetCDF files found in the specified directory.")
```

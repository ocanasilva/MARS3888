## Download NetCDF files

```python
# IMPORT LIBRARIES

import os
import glob
import xarray as xr
from natsort import natsorted

# Import functions (that I am providind to you)
from function_downloads import (
    download_netcdf_ereefs,
    download_netcdf_noaa,
    download_netcdf_csiro
)

# Inform your folder for the downloaded files
download_dir = "downloads"
os.makedirs(download_dir, exist_ok=True)
```

```python
# Download eReefs files

download_netcdf_ereefs(
    base_url="https://thredds.ereefs.aims.gov.au/thredds/fileServer/ereefs/gbr1_2.0/daily-daily",
    download_dir=download_dir,
    start_year=2017, start_month=8, start_day=25,
    end_year=2017, end_month=8, end_day=26,
    file_prefix="EREEFS_AIMS-CSIRO_gbr1_2.0_hydro_daily-daily-",
    date_format="%Y-%m-%d"
)
```

```python
# Download NOAA files

download_netcdf_noaa(
    base_url='https://www.star.nesdis.noaa.gov/pub/socd/mecb/crw/data/5km/v3.1_op/nc/v1.0/daily/dhw/',
    download_dir=download_dir,
    start_year=2020, start_month=3, start_day=1,
    end_year=2020, end_month=3, end_day=6,
    file_prefix='ct5km_dhw_v3.1_',
    date_format='%Y%m%d'
)
```

```python
# Download CAWCR-CSIRO files

download_netcdf_csiro(
    base_url="https://data-cbr.csiro.au/thredds/fileServer/catch_all/CMAR_CAWCR-Wave_archive/CAWCR_Wave_Hindcast_1979-2010/gridded/",
    download_dir=download_dir,
    start_year=2010,
    start_month=1,
    end_year=2010,
    end_month=2,
    file_prefix="ww3.aus_4m.",
    date_format="%Y%m"
)
```

```python
# Read .nc file

# Open a sample NetCDF file
data_files = natsorted(glob.glob(os.path.join(download_dir, "*.nc")))
print("NetCDF files found:")

for f in data_files:
    print(f)

# Create a dataset
data = xr.open_dataset(data_files[1], engine='netcdf4', decode_cf=True)

# List all variable names
print("Variables in the NetCDF file:")
print(list(data.variables))

# Display basic information about the dataset
print("\nDataset info:")
print(data)

# Check additional details like dimensions
print("\nDataset dimensions:")
print(data.dims)

# Print the variable names to inspect further
print("\nAvailable variables:")
for var_name in data.variables:
    print(f"{var_name}: {data[var_name].attrs}")
```
